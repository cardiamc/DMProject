```python
baskets = df.values.tolist()
```

Extract Frequent Patterns

```python
from fim import apriori
```

```python
help(apriori)
```

Looking for all the itemsets with min support of 20%

```python
itemset = apriori(baskets, supp=20, target='a',zmin=2)
print('Number of rule:', len(itemset))
```

```python
baskets
```

We will get a list of tapples, where th efirst component is the pattern self, the secondo component is the support count

```python
itemset[:10]
```

Looking for rules

```python
rules = apriori(baskets, supp=10, zmin=2, target='r', conf=60, 
                report='ascl') 
```

```python
print('Number of rule:', len(rules))
```

```python
for r in rules:
    if r[2] == 'Churn':
        print(r)
```

```python
for r in rules:
    if r[5] > 5 and r[4]> 0.8:
        print(r)
```
