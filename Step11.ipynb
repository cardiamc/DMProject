```python
baskets = df.values.tolist()
```

```python
new_basket=[]

for x in baskets:
    new_list=[]
    for y in x:
        if str(y) != 'nan':
            new_list.append(y)
    new_basket.append(new_list)
```

Extract Frequent Patterns

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
te_ary = te.fit(new_basket).transform(new_basket)
df_trans = pd.DataFrame(te_ary, columns=te.columns_)
df_trans.head()
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
pd.set_option('display.max_rows', 500)
```

```python
frequent_itemsets = apriori(df_trans, min_support=0.15, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets[frequent_itemsets.length>1].sort_values(by='support',ascending=False).reset_index(drop=True)
```
